```python
def solution(dartResult):
    nums = []
    cal_nums = []
    answer = 0
    
    for i in dartResult:
        if(i.isdigit()): 
            nums.append(int(i))
            dartResult = dartResult.replace(i,"n")

    calculation = dartResult.split("n")
    calculation = [i for i in calculation if i]

    for i in range(len(nums)):
        if(nums[i] == 1 and nums[i+1] == 0):
            nums[i] = 10
            nums[i+1] = 'x'

    remove_set = {'x'}
    nums = [i for i in nums if i not in remove_set]

    for i in range(len(nums)):
        if('S' in calculation[i]):
            cal_nums.append(nums[i])
            if('#' in calculation[i]):
                cal_nums[i] *= -1
            elif('*' in calculation[i]):
                if(i == 0):
                    cal_nums[i] *= 2
                else:
                    cal_nums[i] *= 2
                    cal_nums[i-1] *= 2

        elif('D' in calculation[i]):
            cal_nums.append(nums[i] * nums[i])
            if('#' in calculation[i]):
                cal_nums[i] *= -1
            elif('*' in calculation[i]):
                if(i == 0):
                    cal_nums[i] *= 2
                else:
                    cal_nums[i] *= 2
                    cal_nums[i-1] *= 2

        elif('T' in calculation[i]):
            cal_nums.append(nums[i] * nums[i] * nums[i])
            if('#' in calculation[i]):
                cal_nums[i] *= -1
            elif('*' in calculation[i]):
                if(i == 0):
                    cal_nums[i] *= 2
                else:
                    cal_nums[i] *= 2
                    cal_nums[i-1] *= 2
    
    answer = cal_nums[0] + cal_nums[1] + cal_nums[2]
    
    return answer
```
