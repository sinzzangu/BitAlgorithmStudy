```python
def solution(s):
    
    all_count=[]
    for n in range((len(s)//2)+1):
        s_copy=s
        
        print(n)
        
        num_n_str=[]
        c_num=0

        while len(s_copy) > n+1:
            if s_copy[n+1:].startswith(s_copy[:n+1]):
                c_num+=1
                
                print('c_cum=',c_num)
                print(s_copy)

            elif c_num==0:
                num_n_str.append(s_copy[:n+1])

                print(s_copy)
                
            else:
                num_n_str.append(str(c_num+1))
                c_num=0
                num_n_str.append(s_copy[:n+1])
                
                print(s_copy)

            s_copy=s_copy[n+1:]

        else:
            if c_num != 0:
                num_n_str.append(str(c_num+1))
                c_num=0
            num_n_str.append(s_copy)
            
        print(num_n_str)
        
        join_str=''
        for i in range(len(num_n_str)):
            if num_n_str[i].isdigit():
                join_str += str(num_n_str[i])
            else:
                join_str += num_n_str[i]
                
        all_count.append(len(join_str))
    return min(all_count)
        
```

```python
solution("ababcdcdababcdcd")
```
