```python
# first version


ef solution(n, k, cmd):
    deleted=[]
    name=list(range(n))
    
    for i in range(len(cmd)):
        if cmd[i].startswith('U'):
            k-=int(cmd[i][2:])
            
        elif cmd[i].startswith('D'):
            k+=int(cmd[i][2:])

        elif cmd[i]=='C':
            deleted.append([name[k],k])
            del name[k]
            name.append("none")
            if name[k]=='none':
                k-=1

        elif cmd[i]=='Z':
            del name[-1]
            name[deleted[-1][1]+1:]=name[deleted[-1][1]:]
            name[deleted[-1][1]]=deleted[-1][0]
            if deleted[-1][1]<=k:
                k+=1
            del deleted[-1]

    answer=''
    countnum=0
    for i in range(n):
        if i==name[countnum]:
            answer+='O'
            countnum+=1
        else:
            answer+='X'

            
    return answer
```

```python
print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z"]))
print()

print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]))
```

```python
# 2nd version

def solution(n, k, cmd):
    deleted=[]
    answer=list('O'*n)
    
    for i in cmd:
        if i.startswith('U'):
            for j in range(int(i[2:])):
                k-=1
                while answer[k]=='X':
                    k-=1
            
        elif i.startswith('D'):
            for j in range(int(i[2:])):
                k+=1
                while answer[k]=='X':
                    k+=1

        elif i=='C':
            deleted.append(k)
            answer[k]='X'
            if k!=n-1:
                k+=1
                while answer[k]=='X':
                    k+=1
            else:    
                n-=1
                while answer[n-1]=='X':
                    n-=1
                k=n-1
            
        elif i=='Z':
            answer[deleted[-1]]='O'
            if deleted[-1]>n-1:
                n=deleted[-1]+1
            del deleted[-1]
    
    answer=''.join(answer)

    return answer
```

```python
print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z"]))
print()

print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]))
```

```python
# 3rd version

def solution(n, k, cmd):
    deleted=[]
    answer=list('O'*n)
    
    for i in cmd:
        if i.startswith('U'):
            k_b=k
            k-=int(i[2:])
            while answer[k:k_b].count('X')!=0 : 
                num=answer[k:k_b].count('X')
                k_b=k
                k-=num
                 
        elif i.startswith('D'):
            k_b=k
            k+=int(i[2:])
            while answer[k_b+1:k+1].count('X')!=0 :
                num=answer[k_b+1:k+1].count('X')
                k_b=k
                k+=num

        elif i=='C':
            deleted.append(k)
            answer[k]='X'
            if k!=n-1:
                k+=1
                while answer[k]=='X':
                    k+=1
            else:    
                n-=1
                while answer[n-1]=='X':
                    n-=1
                k=n-1
            
        elif i=='Z':
            answer[deleted[-1]]='O'
            if deleted[-1]>n-1:
                n=deleted[-1]+1
            del deleted[-1]
    
    answer=''.join(answer)

    return answer
```

```python
print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z"]))
print()

print(solution(8,2,["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]))
```

```python
# 4th version

def solution(n, k, cmd):
    answer=list(['O']*n)
    
    u_cursor=['null']+list(range(n-1))
    d_cursor=u_cursor[2:]+[n-1,'null']
    
    deleted=[]
    
    for i in cmd:
        if i.startswith('U'):
            for j in range(int(i[2:])):
                k=u_cursor[k]
                 
        elif i.startswith('D'):
            for j in range(int(i[2:])):
                k=d_cursor[k]
            
            
        elif i=='C':
            deleted.append(k)
            answer[k]='X'
            if u_cursor[k]=='null':
                u_cursor[d_cursor[k]]=u_cursor[k]
            elif d_cursor[k]=='null':
                d_cursor[u_cursor[k]]=d_cursor[k]
            else:
                d_cursor[u_cursor[k]]=d_cursor[k]
                u_cursor[d_cursor[k]]=u_cursor[k]

            if k!=n-1:
                k=d_cursor[k]
                
            else:    
                k=u_cursor[k]
                n=k+1
            
        elif i=='Z':
            answer[deleted[-1]]='O'
            if d_cursor[deleted[-1]]=='null':
                d_cursor[u_cursor[deleted[-1]]]=k
            elif u_cursor[deleted[-1]]=='null':
                u_cursor[d_cursor[deleted[-1]]]=k    
            else:
                d_cursor[u_cursor[deleted[-1]]]=k
                u_cursor[d_cursor[deleted[-1]]]=k

            if deleted[-1]>n-1:
                n=deleted[-1]+1
            del deleted[-1]
            
            
        #print(i[0],'k=',k,answer)
    
    answer=''.join(answer)

    return answer
```
